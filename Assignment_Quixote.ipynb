# **Question**

Q: Pick 10 random companies from the entire group and visualise the following through a line chart:

Find the Intrinsic Value of the company based on 3 cases of growth (g): Assume
3 Cases for g (Growth) : Good (15% Growth) ; Bad (-5% Growth); Best (25%Growth)


*   Visualise these 10 companies on a line chart for all 3 cases of growth going
forward against its current market cap.

From Investopedia:

Current Yield = Cashflow/ Market Cap

```python
df_freq['Current Yield'] = df['Operating Cash Flow']/df['Market Cap']
df_freq['Current Yield'].max()
```

 V = (EPS * (8.5 + 2g) * 6) / Y

V : Intrinsic Value

EPS : The Company’s last 12 month earnings per share

8.5 : The constant represents the appropriate P-E ratio for a no-growth
company as proposed by Graham.

g : The company's long-term (five years) earnings growth estimate

6 : The average Return of FDs (6%)

Y : The current yield on AAA corporate bonds.

```python
df_freq['Intrinsic_Value_15_Growth(Good)']= (df['Earnings Per Share']*(8.5 + 2*15)*6)/df_freq['Current Yield']
df_freq['Intrinsic_Value_-5_Growth(Bad)']= (df['Earnings Per Share']*(8.5 + 2*(-5))*6)/df_freq['Current Yield']
df_freq['Intrinsic_Value_25_Growth(Best)']= (df['Earnings Per Share']*(8.5 + 2*25)*6)/df_freq['Current Yield']
```

```python
import random
```

Randomly Picked 10 Companies

```python
Name = df_freq['Name'].sample(n=10)
```

```python
good_growth = df_freq['Intrinsic_Value_15_Growth(Good)'].sample(n=10)
```

```python
bad_growth = df_freq['Intrinsic_Value_-5_Growth(Bad)'].sample(n=10)
```

```python
best_growth = df_freq['Intrinsic_Value_25_Growth(Best)'].sample(n=10)
```

```python

plt.figure(dpi=200)
plt.figure(figsize=(20,8))
x = Name
a = good_growth
b = bad_growth
c= best_growth
# Plot a simple line chart
plt.plot(x, a,'y', label='Good Growth= 15%')
plt.plot(x, b,'r', label='Good Growth=-05%')
# Plot another line on the same chart/graph
plt.plot(x, c, 'g', label='Best Growth=25%')
plt.xlabel('Companies')
plt.ylabel('Companies Growth in Crores')
plt.legend()
plt.show() 
```
